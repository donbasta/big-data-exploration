## Milestone 1
Kerjakan Milestone 1 pada HandsOn Week 6(sebelumnya), akan tetapi menggunakan RDD Spark. Catat waktu (bandingkan) yang dibutuhkan (dalam detik) antara: "MapReduce menggunakan hadoop streaming jar" dengan yang akan kamu proses menggunakan RDD Spark ini.

````python
from time import time

# Tuliskan code kamu di sini

def get_item_and_price_columns(line):
    columns = line.split('\t')
    return (columns[3], float(columns[4]))

def filter_items(columns):
    return columns[0] in ["Consumer Electronics", "Toys"]

time_start = time()

purchases_data = sc.textFile("hdfs://localhost:9000/purchases/purchases.txt")
item_and_price = purchases_data.map(get_item_and_price_columns)
item_price_filter = item_and_price.filter(filter_items)
item_price_sum = item_price_filter.reduceByKey(lambda x, y : x + y)
print(item_price_sum.collect())

time_end = time()
time_ellapsed_milestone_1 = time_end - time_start

print("Waktu yang diperlukan dengan Hadoop:", 86, "detik")
print("Waktu yang diperlukan dengan RDD Spark:", time_ellapsed_milestone_1, "detik")
````

## Milestone 2
Kerjakan Milestone 2 pada HandsOn Week5 (sebelumnya), akan tetapi menggunakan RDD Spark. Catat waktu (bandingkan) yang dibutuhkan (dalam detik) antara: "MapReduce menggunakan hadoop streaming jar" dengan yang akan kamu proses menggunakan RDD Spark ini.

````python
# Tuliskan code kamu di sini

def get_columns(line):
    columns = line.split('\t')
    return (columns[2], (float(columns[4]), columns[3]))

def filter_by_key(columns):
    return columns[0] in ['Atlanta', 'Miami', 'San Francisco']

def get_max_price(a, b):
    if a[0] >= b[0]:
        return a
    return b

time_start = time()

purchases_data = sc.textFile("hdfs://localhost:9000/purchases/purchases.txt")
rdd1 = purchases_data.map(get_columns)
rdd2 = rdd1.filter(filter_by_key)
rdd3 = rdd2.reduceByKey(get_max_price)
rdd4 = rdd3.map(lambda x : (x[0], x[1][0], x[1][1]))
print(rdd4.collect())

time_end = time()
time_ellapsed_milestone_2 = time_end - time_start

print("Waktu yang diperlukan dengan Hadoop:", 99, "detik")
print("Waktu yang diperlukan dengan RDD Spark:", time_ellapsed_milestone_2, "detik")
````

## Milestone 3
Kerjakan Milestone 3 pada HandsOn Week5 (sebelumnya), akan tetapi menggunakan RDD Spark. Catat waktu (bandingkan) yang dibutuhkan (dalam detik) antara: "MapReduce menggunakan hadoop streaming jar" dengan yang akan kamu proses menggunakan RDD Spark ini.

````python
# Tuliskan code kamu di sini

def get_transaction_time(line):
    columns = line.split('\t')
    return columns[1]

def filter_by_time_interval(t):
    return ("09:01" <= t <= "10:00") or ("10:01" <= t <= "11:00")

def add_time_interval(t):
    if ("09:01" <= t <= "10:00"):
        return ("09:01-10:00", 1)
    elif ("10:01" <= t <= "11:00"):
        return ("10:01-11:00", 1)

time_start = time()

purchases_data = sc.textFile("hdfs://localhost:9000/purchases/purchases.txt")
rdd1 = purchases_data.map(get_transaction_time)
rdd2 = rdd1.filter(filter_by_time_interval)
rdd3 = rdd2.map(add_time_interval)
rdd4 = rdd3.reduceByKey(lambda x, y : x + y)
print(rdd4.collect())

time_end = time()
time_ellapsed_milestone_3 = time_end - time_start

print("Waktu yang diperlukan dengan Hadoop:", 92, "detik")
print("Waktu yang diperlukan dengan RDD Spark:", time_ellapsed_milestone_3, "detik")
````

## Milestone 4
Milestone ini dibagi menjadi 4.1, 4.2 dan 4.3 yang masing-masing secara berturut-turut adalah mengerjakan ulang Milestone 1, 2 dan 3 di atas (menggunakan RDD Spark), akan tetapi menggunakan trik "**persist() RDD**" untuk mempercepat prosesnya. Kamu bisa melakukan "**persist**" untuk RDD mana saja yang kamu anggap dapat memberikan waktu proses tercepat.

````python
# Tuliskan code kamu di sini

def get_item_and_price_columns(line):
    columns = line.split('\t')
    return (columns[3], float(columns[4]))

def filter_items(columns):
    return columns[0] in ["Consumer Electronics", "Toys"]

time_start = time()

purchases_data = sc.textFile("hdfs://localhost:9000/purchases/purchases.txt")
item_and_price = purchases_data.map(get_item_and_price_columns).persist()
item_price_filter = item_and_price.filter(filter_items).persist()
item_price_sum = item_price_filter.reduceByKey(lambda x, y : x + y).persist()
print(item_price_sum.collect())

time_end = time()
time_ellapsed_milestone_4_1 = time_end - time_start

print("Waktu Milestone 1:", time_ellapsed_milestone_1, " vs. Waktu Milestone 4.1:", time_ellapsed_milestone_4_1)
````

## Milestone 5
Milestone ini dibagi menjadi 5.1, 5.2 dan 5.3 yang masing-masing secara berturut-turut adalah mengerjakan ulang Milestone 1, 2 dan 3 di atas, akan tetapi menggunakan DataFrame dari Apache Spark. Catat waktu yang diperlukan untuk masing-masing proses (5.1, 5.2 dan 5.3).

````python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("dataframe-spark").getOrCreate()
purchases_df = spark.read.csv("hdfs://localhost:9000/purchases/purchases.txt", inferSchema=True, sep="\t")
purchases_df.show(5)
````

````python
# Tuliskan code kamu di sini
from pyspark.sql.functions import sum

time_start = time()

m1 = purchases_df.select(purchases_df['_c3'], purchases_df['_c4']).filter((purchases_df['_c3'] == "Toys") | (purchases_df['_c3'] == "Consumer Electronics"))
m1.groupBy('_c3').agg(sum('_c4')).show()

time_end = time()

print("Waktu Milestone 5.1:", time_end - time_start)
````

````python
# Tuliskan code kamu di sini
from pyspark.sql.functions import max

time_start = time()

atlanta_max = purchases_df.filter(purchases_df['_c2'] == 'Atlanta').groupBy('_c2').agg(max('_c4').alias('max')).collect()[0].__getitem__('max')
miami_max = purchases_df.filter(purchases_df['_c2'] == 'Miami').groupBy('_c2').agg(max('_c4').alias('max')).collect()[0].__getitem__('max')
sf_max = purchases_df.filter(purchases_df['_c2'] == 'San Francisco').groupBy('_c2').agg(max('_c4').alias('max')).collect()[0].__getitem__('max')

atlanta_max_item = purchases_df.select(purchases_df['_c2'], purchases_df['_c4'], purchases_df['_c3']).filter((purchases_df['_c2'] == 'Atlanta') & (purchases_df['_c4'] == atlanta_max)).limit(1)
miami_max_item = purchases_df.select(purchases_df['_c2'], purchases_df['_c4'], purchases_df['_c3']).filter((purchases_df['_c2'] == 'Miami') & (purchases_df['_c4'] == miami_max)).limit(1)
sf_max_item = purchases_df.select(purchases_df['_c2'], purchases_df['_c4'], purchases_df['_c3']).filter((purchases_df['_c2'] == 'San Francisco') & (purchases_df['_c4'] == sf_max)).limit(1)

atlanta_max_item.union(miami_max_item).union(sf_max_item).show()

time_end = time()

print("Waktu Milestone 5.2:", time_end - time_start)
````

````python
# Tuliskan code kamu di sini
from pyspark.sql.functions import lit, count

time_start = time()

int1 = purchases_df.withColumn("timeInterval", lit("09:01-10:00")).filter((purchases_df['_c1'] >= '09:01') & (purchases_df['_c1'] <= '10:00')).select('timeInterval')
int2 = purchases_df.withColumn("timeInterval", lit("10:01-11:00")).filter((purchases_df['_c1'] >= '10:01') & (purchases_df['_c1'] <= '11:00')).select('timeInterval')

df = int1.union(int2).groupBy("timeInterval").agg(count('*')).show()

time_end = time()

print("Waktu Milestone 5.3:", time_end - time_start)
````
